import random 

class DumbChat:
    """
    Create a list includes different BYD cars on sale
    """

    listA=["BYD, Qin, 2015, Brown",
           "BYD, Song, 2020, Grey", 
           "BYD, Seal, 2023, Bulue",
           "BYD, Yuan, 2013,Black",
          "BYD, Han,2018, White",
          "BYD, Song Plus,2021,Light Grey",
          "BYD,Tang,2023,Brown",
          "BYD,Dolphin,2024,Red"]


    def __init__(self):
        """
        Create attributes usedIndex to store all values which have already appeared
        """
    
        self.usedIndex= set()


    def Answer(self,list):
        """Create a method to return result. 
        If all cars in the list have appeared, the program will return "Above are all BYD cars on sale.".
        If there are still cars not returned as result, the program will choose a car not yet appeared as a result randomly and return to user
        """
        unusedIndex=[i for i in range(len(list)) if i not in self.usedIndex]
        if not unusedIndex:
            return "Above are all BYD cars on sale."
        else:
            b=random.choice(unusedIndex)
            self.usedIndex.add(b)
            return list[b]


    def ask(self):
        """
        The program ask user whether they want to know more BYD cars on sale.
        
        If the user reply "yes",they will get car info on sale.
        
        After 8 times input, the program will finished. 
        
        If user reply is not "yes", print "- Chat end...Have a nice day"
        """
        for i in range(9):
            
            a = input('- Do you want to know more BYD cars on sale? ').upper()
            if a == "YES":
                print('- ' + self.Answer(self.listA))
            else:
                print("- Chat end...Have a nice day")
