```python
#exercise1

# Predefined cities with their weather conditions and costs
cities = {
    "Paris": ("sunny", 200),
    "New York": ("rainy", 150),
    "Tokyo": ("cloudy", 300),
    "Barcelona": ("rainy", 100)
}

# Ask user for their budget
budget = float(input("Enter your budget for the trip (in €): "))

# Iterate through each city to check conditions
for city, (weather, cost) in cities.items():
    if (weather == "sunny" or weather == "cloudy") and cost <= budget:
        print(f"You can visit {city} as the weather is {weather} and the cost is €{cost}.")
    else:
        reason = []
        if weather not in ["sunny", "cloudy"]:
            reason.append("the weather is not suitable")
        if cost > budget:
            reason.append(f"the cost (€{cost}) exceeds your budget")
        
        reasons = " and ".join(reason)
        print(f"You cannot visit {city} because {reasons}.")
```

```python
#exercise2.1
# Define the mood and activity combinations
genre_suggestions = {
    ("happy", "chilling"): "Pop",
    ("happy", "partying"): "Dance",
    ("sad", "studying"): "Jazz",
    ("sad", "chilling"): "Indie",
    ("energetic", "working out"): "Hip-Hop",
    ("energetic", "partying"): "Electronic",
    ("relaxed", "studying"): "Classical",
    ("relaxed", "chilling"): "Acoustic",
}

# Suggested activities based on mood
alternative_activities = {
    "happy": ["partying", "chilling"],
    "sad": ["studying", "chilling"],
    "energetic": ["working out", "partying"],
    "relaxed": ["studying", "chilling"],
}

while True:
    # Get user input for mood and activity
    mood = input("Enter your current mood (happy, sad, energetic, relaxed): ").strip().lower()
    activity = input("Enter your activity (studying, working out, chilling, partying): ").strip().lower()

    # Check if the mood and activity combination exists
    suggested_genre = genre_suggestions.get((mood, activity))

    if suggested_genre:
        print(f"Suggested music genre for your mood ({mood}) and activity ({activity}): {suggested_genre}")
    else:
        print(f"Sorry, no genre suggestion available for the combination of mood ({mood}) and activity ({activity}).")
        
        # Suggest alternative activities based on the current mood
        alternatives = alternative_activities.get(mood)
        if alternatives:
            alternative_string = ", ".join(alternatives)
            print(f"With your mood ({mood}), you could try: {alternative_string}.")
        else:
            print("No alternative activities available for your mood.")

    # Ask if the user wants to continue or exit
    continue_choice = input("Do you want to check another mood and activity? (yes/no): ").strip().lower()
    if continue_choice != "yes":
        print("Thank you! Enjoy your music!")
        break
```

```python
#exercise2.2

def create_quick_workout_plan():
    # Predefined list of exercises and their repetitions
    exercises = {
        "Push-ups": 10,
        "Squats": 15,
        "Jumping jacks": 20,
        "Lunges": 12
    }

    # Ask user for the number of sets
    while True:
        try:
            sets = int(input("How many sets do you want to do for each exercise? "))
            if sets <= 0:
                print("Please enter a positive number for sets.")
            else:
                break
        except ValueError:
            print("Please enter a valid number.")

    # Print the structured workout plan
    print("\nYour Workout Plan:")
    for set_number in range(1, sets + 1):
        print(f"\nSet {set_number}:")
        for exercise, repetitions in exercises.items():
            print(f"{exercise}: {repetitions} repetitions")

# Run the program
create_quick_workout_plan()
```

```python
#exercise3
# Predefined list of exercises
exercises = ["Push-ups", "Squats", "Jumping jacks", "Lunges"]

# Dictionary to store workout plan
workout_plan = {}

print("Welcome to the Workout Plan Generator!")
print("Please choose your exercises from the following list:")

# Display available exercises
for i, exercise in enumerate(exercises, start=1):
    print(f"{i}. {exercise}")

# Collecting user choices
while True:
    try:
        # Prompt user to choose an exercise by entering its number
        choice = int(input("Enter the number of the exercise you'd like to add (or 0 to finish): "))
        
        # Exit loop if user is done choosing exercises
        if choice == 0:
            break
        
        # Validate choice
        if 1 <= choice <= len(exercises):
            chosen_exercise = exercises[choice - 1]
            
            # If the exercise is already in the workout plan, inform the user
            if chosen_exercise in workout_plan:
                print(f"{chosen_exercise} is already in your workout plan.")
                continue
            
            # Ask for sets and repetitions for the chosen exercise
            sets = int(input(f"How many sets of {chosen_exercise} would you like to do? "))
            reps = int(input(f"How many repetitions per set for {chosen_exercise}? "))
            
            # Add exercise details to the workout plan
            workout_plan[chosen_exercise] = {"sets": sets, "reps": reps}
            
        else:
            print("Invalid choice. Please enter a valid number.")
    
    except ValueError:
        print("Please enter a valid number.")

# Display the final workout plan
print("\nHere is your customized workout plan:")
for exercise, details in workout_plan.items():
    print(f"{exercise}: {details['sets']} sets of {details['reps']} reps")

print("Good luck with your workout!")
```

```python
#exercise4
import datetime
import random

def greet_user():
    """Greets the user based on the time of day."""
    current_hour = datetime.datetime.now().hour
    if current_hour < 12:
        print("Good morning!")
    elif 12 <= current_hour < 18:
        print("Good afternoon!")
    else:
        print("Good evening!")

def get_date_time():
    """Prints the current date and time in a readable format."""
    current_datetime = datetime.datetime.now()
    print("Current date and time:", current_datetime.strftime("%Y-%m-%d %H:%M:%S"))

def tell_joke():
    """Prints a random joke from a list of jokes."""
    jokes = [
        "Why did the scarecrow win an award? Because he was outstanding in his field!",
        "Why don’t scientists trust atoms? Because they make up everything!",
        "What do you call fake spaghetti? An impasta!",
        "Why did the bicycle fall over? Because it was two-tired!",
        "What do you call cheese that isn't yours? Nacho cheese!"
    ]
    print(random.choice(jokes))

def main():
    """Main function to interact with the user and execute selected options."""
    while True:
        print("\nPersonal Assistant")
        print("1: Get a greeting")
        print("2: Get the current date and time")
        print("3: Hear a joke")
        print("4: Quit")
        
        choice = input("Enter your choice: ")
        
        if choice == "1":
            greet_user()
        elif choice == "2":
            get_date_time()
        elif choice == "3":
            tell_joke()
        elif choice == "4":
            print("Goodbye!")
            break
        else:
            print("Invalid choice, please try again.")

# Run the main function
if __name__ == "__main__":
    main()
```

```python
#exercise5
# Initialize game variables
treasure_count = 0
game_over = False

# Predefined room contents to simulate randomness
room_contents = ['treasure', 'treasure', 'monster', 'treasure', 'monster']
room_index = 0  # Track which "random" outcome to use

# Game loop
print("Welcome to the Dungeon Adventure!")
print("You are at the entrance of the dungeon. You can explore three rooms.")
print("Be careful! There may be monsters lurking around.")

while not game_over:
    # Player chooses a room to enter
    room_choice = input("\nEnter a room (1, 2, or 3) or type 'exit' to leave the dungeon: ").strip()

    if room_choice == 'exit':
        print("You have chosen to leave the dungeon.")
        break
    elif room_choice in ['1', '2', '3']:
        print(f"\nYou enter Room {room_choice}...")

        # Determine the outcome in the room based on the predefined sequence
        content = room_contents[room_index]
        room_index = (room_index + 1) % len(room_contents)  # Move to the next "random" outcome
        
        if content == 'treasure':
            treasure_count += 1
            print("You found a treasure!")
        elif content == 'monster':
            print("Oh no! A monster has appeared!")
            game_over = True
    else:
        print("Invalid input. Please enter 1, 2, 3, or 'exit'.")
    
    # Display treasure count if game isn't over
    if not game_over:
        print(f"Treasure count: {treasure_count}")

# Game end summary
print("\nGame Over.")
print(f"Final treasure count: {treasure_count}")
```

```python
#exercise6

class Pet:
    def __init__(self, name):
        self.name = name
        self.energy = 50  # Starting energy level
        self.hunger = 50  # Starting hunger level
      
    def play(self):
        self.hunger += 10
        self.energy -= 15
        return self.check_levels()
      
    def feed(self):
        self.hunger -= 15
        if self.hunger < 0:  # Ensure hunger does not go below 0
            self.hunger = 0
        return self.check_levels()

    def rest(self):
        self.energy += 20
        if self.energy > 100:  # Ensure energy does not exceed 100
            self.energy = 100
        return self.check_levels()
        
    def status(self):
        print(f"\n{self.name}'s status:")
        print(f"Energy: {self.energy}")
        print(f"Hunger: {self.hunger}")

    def check_levels(self):
        # Check if the pet is too hungry or too tired
        if self.energy <= 0:
            print(f"{self.name} is too tired! Game over.")
            return False
        if self.hunger >= 100:
            print(f"{self.name} is too hungry! Game over.")
            return False
        return True


def main():
    # Create a pet
    pet_name = input("What is your pet's name? ")
    pet = Pet(pet_name)

    while True:
        action = input("What would you like to do? (play/feed/rest/exit): ").lower()
        if action == "play":
            if pet.play():
                print(f"\nAfter playing with {pet.name}:")
                pet.status()
            else:
                break
        elif action == "feed":
            if pet.feed():
                print(f"\nAfter feeding {pet.name}:")
                pet.status()
            else:
                break
        elif action == "rest":
            if pet.rest():
                print(f"\nAfter letting {pet.name} rest:")
                pet.status()
            else:
                break
        elif action == "exit":
            print("Thanks for taking care of your pet!")
            break
        else:
            print("Invalid action. Please choose play, feed, rest, or exit.")

# Run the program
main()
```
